```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjust the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\Salamh\Desktop\portfolio\movies.csv')


```

```python
# Let's us take a look at the data 

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# filling empty cells in budget and gross with zeros

df['budget'] = df['budget'].fillna(0)

df['gross'] = df['gross'].fillna(0)
```

```python
# Attempting to change data type of columns After filling the empty cells

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Create new column for the date by extracting the year from the released column
df['year_corrected'] = df['released'].astype(str).str.split(',').str[-1].str[:5].str.strip()
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.show()
```

```python
# Let's start to calculate correlation 
df.corr(numeric_only=True, method='pearson') #available methods include: pearson, kendall, spearman
# method is pearson by default, if method para is not specified, then pearson is selected. 
```

```python

correlation_matrix = df.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# if you do df_numerized = df (you will create a reference to the original meaning any change you do to df_numerized
# is always reflected in the original dataset df

df_numerized = df.copy()
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
pd.set_option('display.max_rows', None)
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)
sorted_pairs
```
